```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.25)
```

```python
x
```

```python
y = 4.6 * x + 11.13
```

```python
plt.plot(x,y);
```

```python
def sigmoid(z):
    result = 1 / (1 + np.exp(-z))
    return result
```

```python
y = sigmoid(x)
```

```python
plt.plot(x, y);
```

```python
# relu

def relu(x):
    if x > 0:
        return x
    else:
        return 0
```

```python
y = [relu(a) for a in x]
```

```python
plt.plot(x,y);
```
